We can't impute the missing value of artist_latitude, artist_longitude and song_hotttnesss by using common imputing stradegy since they account a large proportion, which will lead to information distored, negatively influents the model performance, so we drop them

```python
5018/8000,3470/8000
```

```python
train = train.drop(train[train['artist_latitude'].isna()].index)
train = train.drop(train[train['song_hotttnesss'].isna()].index)
```

```python
train.shape
```

We notice that there are also a large proportion of songs with published year unknown. According to the description of the feature 'year', the published year of each song in the dataset is collected from the musicbrainz, where all the information of a song is contributed by community volunteers. Therefore, we interpret that the published year of a song is 0 has certain meanings, for example, some songs are not officially published, they are just spread through a small region. And the difference between 0 and other year values could be consider as the year impact difference on the song hottness, even though this is not precise measured. We keep those rows with year equal 0 

```python
sum(train['year'] == 0) , sum(train['year'] == 0)/train.shape[0]
```

We drop those columns that have only one unique value and are not useful in model fitting

```python
train = train.drop(columns = ['danceability','analysis_sample_rate','energy','genre','artist_hotttnesss','artist_familiarity','artist_name',
                             'artist_location','start_of_fade_out'])
```

## bag-of-words-matrix

```python
tag_string = train.columns.values[-13:-2]
tag_string
```

```python
def combine_tags_title(df, tag_string):
    combine_series = df[tag_string[0]].fillna('').astype(str)
    for i in range(1, len(tag_string)):
        combine_series = combine_series + ' ' + df[tag_string[i]].fillna('').astype(str)

    return combine_series
```

```python
combined_text = combine_tags_title(train, tag_string)
combined_text.head()
```

```python
def string_cleaning(combined_text):
    combined_text = [text.lower() for text in combined_text]
    combined_text = [text[1:] for text in combined_text]
    combined_text = [re.sub(r"[^a-zA-Z\s]", "", text) for text in combined_text]

    return combined_text
```

```python
combined_text = string_cleaning(combined_text)
```

```python
def check_blank(combined_text):
    correct_number = 0
    for text in combined_text:
        word_num = 0
        for i in range(len(text)):
            if text[i] == ' ':
                word_num += 1
                
        word_num = word_num + 1

        if text.count(' ') == word_num - 1:
            correct_number += 1
    return correct_number
```

```python
check_blank(combined_text), train.shape[0]
```

After cleaing all the texts, we can create the bag-of-words matrix

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
word_matrix_sci = vectorizer.fit_transform(combined_text)
```

```python
vectorizer.vocabulary_ #value: column index of the word in the matrix
```

```python
word_matrix = word_matrix_sci.toarray()
```

```python
word_matrix_df = pd.DataFrame(word_matrix)   #keep a df version of word matrix
for name, index in vectorizer.vocabulary_.items():     #rename the column index of word data frame
    word_matrix_df = word_matrix_df.rename(columns = {index: name})
```

```python
column_sum = word_matrix.sum(axis = 0)
sorted_index = np.argsort(-column_sum)
sorted_index
```

```python
column_sum[sorted_index]
```

We keep about the top 30% frequentist words appearing in the song titles or artist tags

In other words, we keep the words that at least have counting number larger than 2

```python
((column_sum == 1).sum() + (column_sum == 2).sum()) / len(column_sum)
```

```python
word_matrix = word_matrix[:,sorted_index]
np.where(word_matrix.sum(axis = 0)<3)
```
