We don't use array type value feature of the song

```python
def single_hdf5_read_analysis(file):
    dic = {'analysis_sample_rate': file['analysis/songs'][0][0],
          'danceability': file['analysis/songs'][0][2],
           'duration': file['analysis/songs'][0][3],
           'end_of_fade_in': file['analysis/songs'][0][4],
           'energy': file['analysis/songs'][0][5],
           'key': file['analysis/songs'][0][21],
           'loudness': file['analysis/songs'][0][23],
           'mode': file['analysis/songs'][0][24],
           'start_of_fade_out': file['analysis/songs'][0][26],
           'tempo': file['analysis/songs'][0][27],
           'time_signature': file['analysis/songs'][0][28],
          }
    return pd.DataFrame(dic, index = [0])
```

```python
def single_hdf5_read_metadata_songs(file):
    dic = {'artist_familiarity': file['metadata/songs'][0][2],
          'artist_hotttnesss': file['metadata/songs'][0][3],
           'artist_latitude': file['metadata/songs'][0][5],
           'artist_location': file['metadata/songs'][0][6],
           'artist_longitude': file['metadata/songs'][0][7],
           'artist_name': file['metadata/songs'][0][9],
           'genre': file['metadata/songs'][0][11],
           'release': file['metadata/songs'][0][14],
           'song_hotttnesss': file['metadata/songs'][0][16],
           'title': file['metadata/songs'][0][18]
          }
    return pd.DataFrame(dic, index = [0])
```

```python
single_hdf5_read_metadata_songs(f)
```

plus
- top 10 artist terms in metadata

```python
def single_hdf5_read_metadata_artist_terms_top10(file):
    lis = []
    lis = [('tag_top_'+str(index+1),tag.astype(str)) for index, tag in enumerate(file['metadata/artist_terms'][:10]) ]
    lis = dict(lis)
    return pd.DataFrame(lis, index=[0])
```

Then, we can combine all the columns we get from the data of a single song

```python
func_list = [single_hdf5_read_analysis, single_hdf5_read_metadata_songs, single_hdf5_read_metadata_artist_terms_top10,
             single_hdf5_read_musicbrainz_year]
```

```python
def comb_columns_single_song(func_list, file_path):
    file = h5py.File(file_path, 'r')
    return pd.concat([func_list[0](file),func_list[1](file),func_list[2](file),func_list[3](file)], axis = 1)
```

We start to extract the relevant data of all the songs

```python
directory = "D:/songdata/MillionSongSubset"
```

```python
file_paths = []
for dirpath, dirnames, filenames in os.walk(directory):
    for filename in filenames:
        if filename.endswith('.h5'):
            file_paths.append(dirpath+'/'+filename)
```

```python
file_paths
```

```python
file_paths = [file_path.replace('\\', '/') for file_path in file_paths]
```

```python
file_paths
```

```python
len(file_paths)
```

```python
range(len(file_paths))
```

```python
df = pd.DataFrame()
for i in range(0,len(file_paths), 2):
    df_in_step = pd.DataFrame()
    df_in_step = pd.concat([comb_columns_single_song(func_list, file_paths[i]), comb_columns_single_song(func_list, file_paths[i+1])], 
                            ignore_index = True)
    df = pd.concat([df, df_in_step], ignore_index = True)
```
